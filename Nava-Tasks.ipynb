```python
from tkinter import *
from tkinter import messagebox

def newTask():
    task = my_entry.get()
    if task != "":
        my_list.insert(END, task)
        my_entry.delete(0, "end")
    else:
        messagebox.showwarning("warning", "Please enter some task.")

def deleteTask():
    my_list.delete(ANCHOR)

def task_complete():
    selected_task_index = my_list.curselection()
    if selected_task_index:
        task_info = my_list.get(selected_task_index)
        task, is_completed = task_info
        updated_task_info = (task, not is_completed)
        my_list.delete(selected_task_index)
        my_list.insert(selected_task_index, updated_task_info)
    else:
        messagebox.showwarning("Warning", "Please select a task to mark as complete/incomplete.")
  
def quit_task():
    root.destroy() 

ws = Tk()
ws.geometry('600x500+500+200')
ws.title('To-Do-List')
ws.config(bg='#EAEAEA')
ws.resizable(width=False, height=False)

frame = Frame(ws)
frame.pack(pady=10)

my_list = Listbox(
    frame,
    width=40,
    height=8,
    font=('Kunstler', 18),
    bd=0,
    fg='#1E1E1E',
    highlightthickness=0,
    selectbackground='#EE82EE',
    activestyle="none",
    
)
my_list.pack(side=LEFT, fill=BOTH)

task_list = [
    'Social media dump',
    'Declutter',
    'write software',
    'Shop groceries',
    'Take a nap',
    'Learn something',
    'Shoot the sherrif'
    ]

for item in task_list:
    my_list.insert(END, item)

sb = Scrollbar(frame)
sb.pack(side=RIGHT, fill=BOTH)

my_list.config(yscrollcommand=sb.set)
sb.config(command=my_list.yview)

my_entry = Entry(
    ws,
    font=('Edwardian', 30),
    width=40
    )

my_entry.pack(pady=20)

button_frame = Frame(ws)
button_frame.pack(pady=20)

addTask_btn = Button(
    button_frame,
    text='Add Task',
    font=('Script 20'),
    bg='#c5f776',
    padx=20,
    pady=10,
    command=newTask
)
addTask_btn.pack(fill=BOTH, expand=True, side=LEFT)

delTask_btn = Button(
    button_frame,
    text='Delete Task',
    font=('Script 20'),
    bg='#ff8b61',
    padx=20,
    pady=10,
    command=deleteTask
)
delTask_btn.pack(fill=BOTH, expand=True, side=LEFT)

complTask_btn = Button(
    button_frame,
    text='Task Complete',
    font=('Script 20'),
    bg='#473C8B',
    padx=20,
    pady=10,
    command=task_complete
)
complTask_btn.pack(fill=BOTH, expand=True, side=LEFT)



QuitTask_btn = Button(
    button_frame,
    text='Quit',
    font=('Script' '40' 'Bold'),
    bg='#FFFF00',
    padx=20,
    pady=10,
    command=quit_task
)
QuitTask_btn.pack(fill=BOTH, expand=True, side=LEFT)

ws.mainloop()
```
